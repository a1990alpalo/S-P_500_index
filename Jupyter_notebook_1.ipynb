```python
#importing libraries to begin the project
import pandas as pd    
```

```python
#assigning a name to the csv file 
file_one = "Resources/sp500_companies.csv"
```

```python
#Now that we have assigned a name to the csv we can load the pandas dataframe 
file_one_df = pd.read_csv(file_one)

file_one_df.head(20)
```

```python
file_one_df.columns
```

```python
#Summary of the dataset// This cell calculates the number of different types of industries 

#Tracking the total count of industry sector/s dataset
industry_count = len(file_one_df["Industry"].unique())
#Tracking the total count of sectors in the dataset
sector_count = len(file_one_df["Sector"].unique())
#Tracking the total count of states where these industries reside 
state_count = len(file_one_df["State"].unique())
#Total number of cities where industries reside 
city_count = len(file_one_df["City"].unique())
```

```python
#Lets create a summary dataframe with all the different sectors in the dataset

summary_df = pd.DataFrame({
    "Total number of industry": [industry_count],
    "Total number of sector": [sector_count],
    "Total number of different states":[state_count],
    "Total number of cities" : [city_count]
    
    })

#Display the summary_df datraframe 
summary_df
```
